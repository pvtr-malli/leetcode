Given the roots of two binary trees root and subRoot, return true if there is a subtree of root with the same structure and node values of subRoot and false otherwise.

A subtree of a binary tree tree is a tree that consists of a node in tree and all of this node's descendants. The tree tree could also be considered as a subtree of itself.

 

Example 1:


Input: root = [3,4,5,1,2], subRoot = [4,1,2]
Output: true
Example 2:


Input: root = [3,4,5,1,2,null,null,null,null,0], subRoot = [4,1,2]
Output: false
 

Constraints:

The number of nodes in the root tree is in the range [1, 2000].
The number of nodes in the subRoot tree is in the range [1, 1000].
-104 <= root.val <= 104
-104 <= subRoot.val <= 104

```python
from typing import Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isSameTree_rec(self, r1: TreeNode, r2: TreeNode):
        if r1 == None or r2 == None:
            return r1 == r2
        
        if r1.val == r2.val:
            right = self.isSameTree_rec(r1.right, r2.right)
            left = self.isSameTree_rec(r1.left, r2.left)
            return right and left 
        else:
            return False

    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        # Find the sub tree root node.
        if not root:
            return False 
        
        if not subRoot:
            return True
        
        q = []
        q.append(root)
        while q:
            node = q.pop(0)
            if node.val == subRoot.val:
                if self.isSameTree_rec(node, subRoot):
                    return True
                # if not, there could be anther node with this value, keep searching
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)

        # if we didn't find the subroot, in the root. return false.
        return False
        
# ✅ Time & Space Complexity:
# Time: Worst-case O(m * n) — where m is the number of nodes in root, and n is number of nodes in subRoot.
# Space: O(max(width of tree)) for BFS queue + O(h) for recursion stack in isSameTree_rec.

```

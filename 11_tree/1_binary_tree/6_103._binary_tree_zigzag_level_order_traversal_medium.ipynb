Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

 

Example 1:


Input: root = [3,9,20,null,null,15,7]
Output: [[3],[20,9],[15,7]]
Example 2:

Input: root = [1]
Output: [[1]]
Example 3:

Input: root = []
Output: []
 

Constraints:

The number of nodes in the tree is in the range [0, 2000].
-100 <= Node.val <= 100

```python
from typing import Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> list[list[int]]:
        if not root:
            return []
        queue = []
        left_to_right = True
        ans = []
        queue.append(root)

        while queue:
            # prepare the list which is gonna store the values.
            q_len = len(queue)
            level_arr = [-1] * q_len
            # prepate the index, if we wanna go right to left.
            index = range(0, q_len) # oyu have to go from 0 - n-1
            if not left_to_right: 
                index = range(q_len-1, -1, -1) # you have to go from n-1 -> 0

            # get all the value from the queue and push their left and right.
            for i in index:
                ele = queue.pop(0)
                level_arr[i] = ele.val

                if ele.left:
                    queue.append(ele.left)

                if ele.right:
                    queue.append(ele.right)
                
            
            ans.append(level_arr)
            # change the direction.
            # left_to_right = False  if left_to_right else True
            left_to_right = not left_to_right # flip the direction
        return ans
    
# tc - O(n)
# sc - O(n) [quueu] + O(n) [ res arr]
```

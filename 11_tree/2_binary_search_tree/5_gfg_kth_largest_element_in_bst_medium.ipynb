Given a Binary Search Tree. Your task is to complete the function which will return the kth largest element without doing any modification in the Binary Search Tree.

Examples:

Input:
      4
    /   \
   2     9
k = 2 
Output: 4
Explanation: 2nd Largest element in BST is 4
Input:
       9
        \ 
          10
k = 1
Output: 10
Explanation: 1st Largest element in BST is 10
Input:
      4
    /   \
   2     9
k = 3 
Output: 2
Explanation: 3rd Largest element in BST is 2
Constraints:
1 <= number of nodes <= 105
1 <= node->data <= 105
1 <= k <= number of nodes

Expected Complexities
Time Complexity: O(n)
Auxiliary Space: O(Height of BST)

```python
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

class Solution:
    def cunt_nodes(self, root:Node):
        if root == None:
            return 0
        left = self.cunt_nodes(root.left)
        right = self.cunt_nodes(root.right)
        return 1 + left + right 

    def kthSmallest(self, root, k):
        # gonna use the ierative inorder solution.
        s = [] # lifo
        count = 0
        while True:
            if root != None:
                s.append(root)
                root = root.left
            else:
                if  len(s) == 0:
                    break
                root = s.pop()
                # we got the root val so count it.
                count+=1
                if count == k:
                    return root.data
                root = root.right
    
    def kthLargest(self, root:Node, k: int):
        # find the count of the node.
        cnt = self.cunt_nodes(root)

        # so getting the kth largest number is == getting the cnt - k th smallest ndoe.
        k_small = (cnt - k + 1) #  (cnt - k)th node (0-based), but kthSmallest() expects a 1-based k.
        return self.kthSmallest(root, k_small)
    
# tc - O(n)
# sc - O(H)
```

You are given the root of a binary search tree (BST), where the values of exactly two nodes of the tree were swapped by mistake. Recover the tree without changing its structure.

 

Example 1:


Input: root = [1,3,null,null,2]
Output: [3,1,null,null,2]
Explanation: 3 cannot be a left child of 1 because 3 > 1. Swapping 1 and 3 makes the BST valid.
Example 2:


Input: root = [3,1,4,null,null,2]
Output: [2,1,4,null,null,3]
Explanation: 2 cannot be in the right subtree of 3 because 2 < 3. Swapping 2 and 3 makes the BST valid.
 

Constraints:

The number of nodes in the tree is in the range [2, 1000].
-231 <= Node.val <= 231 - 1
 

Follow up: A solution using O(n) space is pretty straight-forward. Could you devise a constant O(1) space solution?

```python

class Solution:
    def recoverTree(self, root: Optional[TreeNode]) -> None:
        nodes = []
        
        # In-order traversal to collect nodes
        def inorder(node):
            if not node:
                return
            inorder(node.left)
            nodes.append(node)
            inorder(node.right)

        inorder(root)

        # Identify the two swapped nodes.
        x = y = None
        for i in range(1, len(nodes)):
            # [3,25,8,7,10,15,20,5]
            # whwen this is not following the trend, take that value.... as y....
            # here when i = 2, that is not following the trend.. so y = 8 and x = 25
            # when i = 7, again its not following the trens - y = 7
            
            if nodes[i].val < nodes[i-1].val:
                y = nodes[i]
                if not x:
                    x = nodes[i-1]
                else:
                    break

        # Swap back their values
        x.val, y.val = y.val, x.val

# tc - O(n)
# sc - O(n)
```

```python
# do the violaion check whenever we add the inorder to the ans in morrris traversal.

class Solution:
    def recoverTree(self, root: Optional[TreeNode]) -> None:
        first = second = prev = None
        curr = root

        while curr:
            if not curr.left:
                # we are in the left, do the check.
                # [1,5] - valid and [10,8] - invalid sp invalid when prev > cur.
                if prev and prev.val > curr.val:
                    if not first:
                        first = prev
                    second = curr
                # keep track of 2 values, prev and cur. 
                prev = curr
                curr = curr.right
            else:
                # Find the inorder predecessor.
                # just the morris traversal.
                pre = curr.left
                while pre.right and pre.right != curr:
                    pre = pre.right

                if not pre.right:
                    pre.right = curr
                    curr = curr.left
                else:
                    pre.right = None
                    # do the check
                    if prev and prev.val > curr.val:
                        if not first:
                            first = prev
                        second = curr
                    # track of both the values.
                    prev = curr
                    curr = curr.right

        # Swap the values of the two nodes
        if first and second:
            first.val, second.val = second.val, first.val

# tc - O(N)
# sc - O(1)
```

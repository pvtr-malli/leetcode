Given an array nums of n integers, return an array of all the unique quadruplets [nums[a], nums[b], nums[c], nums[d]] such that:

0 <= a, b, c, d < n
a, b, c, and d are distinct.
nums[a] + nums[b] + nums[c] + nums[d] == target
You may return the answer in any order.

 

Example 1:

Input: nums = [1,0,-1,0,-2,2], target = 0
Output: [[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]
Example 2:

Input: nums = [2,2,2,2,2], target = 8
Output: [[2,2,2,2]]
 

Constraints:

1 <= nums.length <= 200
-109 <= nums[i] <= 109
-109 <= target <= 109

```python
# approach one:
# - sorting, 3 loop + one binary search for target - ( arr[a] + arr[b] + arr[c]) + hash to get the unique values sets.
# tc - O(nlogn)[sorting]+ O(n^3) [looping] + O(log n)[bianry search]
# sc - O(n)[target]
```

```python
# approach two>:
# sorting | 2 loop + 2 poniter [start and end]
# tc - O(nlogn)[sorting]+ O(n^2) [looping] + O(n)[2 poniters]
# sc - O(1)[will skip the duplicates on the way]


# approach two>:
# sorting | 2 loop + 2 poniter [start and end] + early stoppping
# tc - O(nlogn)[sorting]+ O(n^2) [looping] + O(n)[2 poniters] 
# sc - O(1)[will skip the duplicates on the way]
```

```python
class Solution:
    def fourSum(self, nums, target):
        nums.sort()
        n = len(nums)
        res = []

        for i in range(n - 3): # NOTE
            # Skip duplicate i
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            #  Meaning: The smallest possible sum starting from i is already too big.
            # Since the array is sorted, nums[i], nums[i+1], etc. are the smallest possible values from index i.
            # If even the minimum possible sum from this point is greater than target, then no further i will work, so we break the loop.
            if nums[i] + nums[i+1] + nums[i+2] + nums[i+3] > target:
                break

            # Meaning: The largest possible sum from i is still smaller than the target.
            # nums[n-3], nums[n-2], nums[n-1] are the largest values available.
            # If the maximum sum with nums[i] is still less than the target, then i is too small â€” so we continue to the next i.
            
            if nums[i] + nums[n-3] + nums[n-2] + nums[n-1] < target:
                continue

            for j in range(i + 1, n - 2): # NOTE:
                # Skip duplicate j
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue

                # Early stop
                if nums[i] + nums[j] + nums[j+1] + nums[j+2] > target:
                    break

                # Early continue
                if nums[i] + nums[j] + nums[n-2] + nums[n-1] < target:
                    continue

                left, right = j + 1, n - 1

                while left < right:
                    total = nums[i] + nums[j] + nums[left] + nums[right]

                    if total == target:
                        res.append([nums[i], nums[j], nums[left], nums[right]])

                        # Move and skip duplicate
                        left += 1
                        right -= 1

                        while left < right and nums[left] == nums[left - 1]:
                            left += 1
                        while left < right and nums[right] == nums[right + 1]:
                            right -= 1

                    elif total < target:
                        left += 1
                    else:
                        right -= 1

        return res
```

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.
 

Constraints:

3 <= nums.length <= 3000
-105 <= nums[i] <= 105

```python
# one for loop and a 2 pointer approach like in 3ed sum.


class Solution:
    def threeSum(self, nums):
        
        nums = sorted(nums)

        # base conditions
        length = len(nums)
        if length <= 2 :
            return []
        
        threesum_set = []
        for a in range(0, length - 2):
            # for each we gonna run the 2 pointers (to fine the triplets)
            if a == 0 or (a > 0 and nums[a] != nums[a - 1]): # skipping the duplicates
                low = a + 1
                high = length - 1
                wanted_sum = -nums[a]

                while low < high:
                    if nums[low] + nums[high] == wanted_sum:
                        threesum_set.append([nums[a], nums[low], nums[high]])

                        # skipping the duplicates.
                        while (low < high and nums[low] == nums[low + 1]):
                            low += 1
                        while (low > high and nums[high] == nums[high + 1]):
                            high -= 1

                        low +=1 ; high -= 1
                    elif nums[low] + nums[high] < wanted_sum:
                        low += 1
                    else:
                        high -=1 
        return threesum_set
    

# tc - O(nlong) + O(n^2) = O(n^2)
# sc - O(#ans)[output array]
```

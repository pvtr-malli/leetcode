You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

 

Example 1:


Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
Example 2:

Input: height = [1,1]
Output: 1

see diagram: https://leetcode.com/problems/container-with-most-water/description/

````python
class Solution: # this is for trapping rain water.
    def maxArea(self, height):
        # try to keep both max, if something is getting small move the pointer.
        area,i,j=0,0,len(height)-1
        left_max = 0
        right_max = 0

        while i < j:
            if height[i] < height[j]:
                if height[i] > left_max:
                    left_max = height[i]
                else:
                    # IF I am hear then something beigger than that is in the right side.
                    # so taking care of only this bar is enough.
                    area += left_max- height[i]
                i += 1
            else:
                if height[j] > right_max:
                    right_max = height[j]
                else:
                    area += right_max- height[j]
                j-=1
        return area
````



You are given an array `height`:

```
height[i]` = height of vertical line at position i
```

You want to **find two lines that, together with the x-axis, can contain the most water**.

**Area between two lines** =

[
\text{Area} = \text{width} \times \text{height of shorter line} = (j - i) * min(height[i], height[j])
]

---

# ðŸ§  Logic / Approach

We use a **two-pointer approach**:

1. Start with **two pointers**, `i` at the start (0) and `j` at the end (n-1).
2. Calculate the **area between them**:

```python
area = min(height[i], height[j]) * (j - i)
```

3. Keep track of **maximum area found** so far:

```python
max_area = max(max_area, area)
```

4. **Move the pointer pointing to the shorter line**:

* Why? Because moving the taller line **cannot increase the area**, since height is limited by the shorter line.
* Moving the shorter line **might find a taller line**, potentially increasing area.

5. Repeat until `i < j`.

---




````python
class Solution: # this is for the container with most water. 

    def maxArea(self, height):
        # try to keep both max, if something is getting small move the pointer.
        max_area,i,j=0,0,len(height)-1
        while i<j:
            # The area is limited by the minimum point. so find hte area based on the minimum.
            # area = height * width 
            cur_area = min(height[i], height[j]) * (j - i)
            max_area = max(max_area, cur_area)
            if height[i]<height[j]:
                i+=1
            else:
                j-=1
        return max_area
````

````python
Solution().maxArea([1,8,6,2,5,4,8,3,7])
````

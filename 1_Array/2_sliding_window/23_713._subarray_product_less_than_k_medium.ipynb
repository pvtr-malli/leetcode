Given an array of integers nums and an integer k, return the number of contiguous subarrays where the product of all the elements in the subarray is strictly less than k.

 

Example 1:

Input: nums = [10,5,2,6], k = 100
Output: 8
Explanation: The 8 subarrays that have product less than 100 are:
[10], [5], [2], [6], [10, 5], [5, 2], [2, 6], [5, 2, 6]
Note that [10, 5, 2] is not included as the product of 100 is not strictly less than k.
Example 2:

Input: nums = [1,2,3], k = 0
Output: 0
 

Constraints:

1 <= nums.length <= 3 * 104
1 <= nums[i] <= 1000
0 <= k <= 106

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums, k):
        count = 0
        n = len(nums)

        for i in range(n):
            product = nums[i]
            if product < k:
                count += 1
            for j in range(i + 1, n):
                product *= nums[j]
                if product >= k:
                    break
                count += 1

        return count
# This is a brute-force approach.

# For each starting index i, it tries to extend the subarray to the right (j) until the product exceeds or equals k.

# Time complexity: O(n²) in the worst case.
```

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:        
        start = 0
        res = 0
        product = 1
        
        # increase the window. 
        for end in range(len(nums)):
            product *= nums[end]
            
            # decrease the window, until it's not violating the condition.
            if product >= k:
                while product >= k and start <= end:
                    product /= nums[start]
                    start += 1
            
            # tracker for the current window.

            # why the res is this,
            # All the subarrys inside this window are having prod < k. since the whole subarray prod < k.
            # so for the current ind, calculate how many subarrys are in the window having right as the last ind.
            # start = 1
            # end = 3
            # Subarray: nums[1:4] = [5, 2, 6]
            # Product = 60 < 100 ✅
            # Now valid subarrays ending at index 3:
            # [5, 2, 6]
            # [2, 6]
            # [6]
            # this is nothing but the number of elements in the window. so end - right + 1
            res += end - start + 1
        
        return res
    
# O(n)
# O(1)
```

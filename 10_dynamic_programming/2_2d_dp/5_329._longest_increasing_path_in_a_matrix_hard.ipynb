```python
# memorization:
class Solution:
    def longestIncreasingPath(self, matrix: list[list[int]]) -> int:
        n, m = len(matrix), len(matrix[0])
        memo = [[-1 for _ in range(m)] for _ in range(n)]  # store longest path starting at (i,j)
        
        dx_arr = [0, 0, -1, 1]  # left, right, up, down
        dy_arr = [-1, 1, 0, 0]
        
        def dfs(i, j):
            if memo[i][j] != -1:
                return memo[i][j]
            
            max_len = 1  # at least the cell itself
            for dx, dy in zip(dx_arr, dy_arr):
                ni, nj = i + dx, j + dy
                if 0 <= ni < n and 0 <= nj < m and matrix[ni][nj] > matrix[i][j]:
                    max_len = max(max_len, 1 + dfs(ni, nj))
            
            memo[i][j] = max_len
            return max_len
        
        result = 0
        for i in range(n):
            for j in range(m):
                result = max(result, dfs(i,j))
        
        return result


# tc:
# - Outer loop calls dfs(i,j) for each of n×m cells
# BUT each dfs(i,j) is computed EXACTLY ONCE across ALL calls!

# First call to dfs(i,j): Does actual work
# Subsequent calls to dfs(i,j): Return cached result in O(1)

# Total work done = Number of unique subproblems × Work per subproblem

# Unique subproblems: n×m (one per cell)
# Work per subproblem: O(1) (check 4 neighbors)
# So Total TC: O(n×m)

# sc - O(n * m)
```

Yes, the **optimal approach** for Longest Increasing Path in a Matrix is **DFS + memoization**, which we already discussed.

* **Tabulation / bottom-up DP** is tricky here because the matrix cells depend on neighbors in any direction, and you don’t know a valid order to fill the DP array without first topologically sorting the cells.
* Effectively, the DFS + memoization acts like **topological DP**: each cell is computed only once, and dependencies are resolved recursively.
* So **DFS + memoization is considered the optimal solution**, with **O(n × m)** time and space.

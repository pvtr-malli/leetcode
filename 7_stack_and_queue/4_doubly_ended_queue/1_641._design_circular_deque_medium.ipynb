Design your implementation of the circular double-ended queue (deque).

Implement the MyCircularDeque class:

MyCircularDeque(int k) Initializes the deque with a maximum size of k.
boolean insertFront() Adds an item at the front of Deque. Returns true if the operation is successful, or false otherwise.
boolean insertLast() Adds an item at the rear of Deque. Returns true if the operation is successful, or false otherwise.
boolean deleteFront() Deletes an item from the front of Deque. Returns true if the operation is successful, or false otherwise.
boolean deleteLast() Deletes an item from the rear of Deque. Returns true if the operation is successful, or false otherwise.
int getFront() Returns the front item from the Deque. Returns -1 if the deque is empty.
int getRear() Returns the last item from Deque. Returns -1 if the deque is empty.
boolean isEmpty() Returns true if the deque is empty, or false otherwise.
boolean isFull() Returns true if the deque is full, or false otherwise.
 

Example 1:

Input
["MyCircularDeque", "insertLast", "insertLast", "insertFront", "insertFront", "getRear", "isFull", "deleteLast", "insertFront", "getFront"]
[[3], [1], [2], [3], [4], [], [], [], [4], []]
Output
[null, true, true, true, false, 2, true, true, true, 4]

Explanation
MyCircularDeque myCircularDeque = new MyCircularDeque(3);
myCircularDeque.insertLast(1);  // return True
myCircularDeque.insertLast(2);  // return True
myCircularDeque.insertFront(3); // return True
myCircularDeque.insertFront(4); // return False, the queue is full.
myCircularDeque.getRear();      // return 2
myCircularDeque.isFull();       // return True
myCircularDeque.deleteLast();   // return True
myCircularDeque.insertFront(4); // return True
myCircularDeque.getFront();     // return 4
 

Constraints:

1 <= k <= 1000
0 <= value <= 1000
At most 2000 calls will be made to insertFront, insertLast, deleteFront, deleteLast, getFront, getRear, isEmpty, isFull.

# brute force:
- simulate it using the python list.

```python
class MyCircularDeque:

    def __init__(self, k: int):
        self.k = k
        self.deque = []

    def insertFront(self, value: int) -> bool:
        if len(self.deque) < self.k:
            self.deque.insert(0, value)
            return True
        return False

    def insertLast(self, value: int) -> bool:
        if len(self.deque) < self.k:
            self.deque.append(value)
            return True
        return False

    def deleteFront(self) -> bool:
        if self.deque:
            self.deque.pop(0)
            return True
        return False

    def deleteLast(self) -> bool:
        if self.deque:
            self.deque.pop()
            return True
        return False

    def getFront(self) -> int:
        return self.deque[0] if self.deque else -1

    def getRear(self) -> int:
        return self.deque[-1] if self.deque else -1

    def isEmpty(self) -> bool:
        return len(self.deque) == 0

    def isFull(self) -> bool:
        return len(self.deque) == self.k


# | Operation       | Time |
# | --------------- | ---- |
# | `insertFront()` | O(n) |
# | `insertLast()`  | O(1) |
# | `deleteFront()` | O(n) |
# | `deleteLast()`  | O(1) |
# | `getFront()`    | O(1) |
# | `getRear()`     | O(1) |
# | `isEmpty()`     | O(1) |
# | `isFull()`      | O(1) |

# sc - O(k) â€” max size of the deque is k.
```

# lets make the insert and delete also O(1) using linked list.

```python
# page 44:

class Node:
    def __init__(self, val=None):
        self.val = val
        self.prev = None
        self.next = None

class MyCircularDeque:
    def __init__(self, k: int):
        self.k = k
        self.size = 0

        self.head = Node()  # dummy head
        self.tail = Node()  # dummy tail

        self.head.next = self.tail
        self.tail.prev = self.head

    def _add(self, prev, new_node):
        nxt = prev.next

        prev.next = new_node
        new_node.prev = prev
        
        new_node.next = nxt
        nxt.prev = new_node

    def _del(self, node):
        prev = node.prev
        nxt = node.next
        prev.next = nxt
        nxt.prev = prev

    def insertFront(self, value: int) -> bool:
        if self.size == self.k:
            return False
        self._add(self.head, Node(value))
        self.size += 1
        return True

    def insertLast(self, value: int) -> bool:
        if self.size == self.k:
            return False
        self._add(self.tail.prev, Node(value))
        self.size += 1
        return True

    def deleteFront(self) -> bool:
        if self.size == 0:
            return False
        self._del(self.head.next)
        self.size -= 1
        return True

    def deleteLast(self) -> bool:
        if self.size == 0:
            return False
        self._del(self.tail.prev)
        self.size -= 1
        return True

    def getFront(self) -> int:
        return self.head.next.val if self.size > 0 else -1

    def getRear(self) -> int:
        return self.tail.prev.val if self.size > 0 else -1

    def isEmpty(self) -> bool:
        return self.size == 0

    def isFull(self) -> bool:
        return self.size == self.k


# tc - O(1) for all operations.
# sc - O(k)
```

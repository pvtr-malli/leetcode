# Brtute force:
- flattent the list in prior then just call the next and hasnext
- flattent - O(n)
- hasnext() and next() - O(1)
- sc - O(n)


````python
class NestedIterator:
    def __init__(self, nestedList):
        self.flat_list = []
        self.index = 0
        self.flatten(nestedList)
        print(self.flat_list)

    def flatten(self, nestedList):
        for item in nestedList:
            if item.isInteger():
                self.flat_list.append(item.getInteger())
            else:
                print(item.getList())
                self.flatten(item.getList())

    def next(self) -> int:
        val = self.flat_list[self.index]
        self.index += 1
        return val

    def hasNext(self) -> bool:
        return self.index < len(self.flat_list)

# this is the output from hte leedcode for the print statements:
# [NestedInteger{_integer: 1, _list: []}, NestedInteger{_integer: 1, _list: []}]
# [NestedInteger{_integer: 1, _list: []}, NestedInteger{_integer: 1, _list: []}]
# [1, 1, 2, 1, 1]

# so we are recursively flattening the lists.
````

# Approach 2:
- instead of flattening in the beggining itself, what we can do is. while calling list 
- put all the values in the stack - O(n)
```

while stack:
    if the top is interger:

        we pop it
        put it in the stack one by one in reversed order.
    return False
```

````python
class NestedIterator:
    def __init__(self, nestedList):
        self.stack = nestedList[::-1]

    def next(self) -> int:
        return self.stack.pop().getInteger()

    def hasNext(self) -> bool:
        while self.stack:
            top = self.stack[-1]
            if top.isInteger():
                return True
            self.stack.pop()
            self.stack.extend(top.getList()[::-1])
        return False


# next - O(1) always
# hasNext(): Worst-case O(N) in total across all calls, where N is the total number of nested elements.
# Why? Each nested list is unpacked only once, and each element is processed only once. 
# So even though hasNext() may appear to do a loop, the total number of iterations across the entire use of the iterator is still linear.

# sc - O(n)
````

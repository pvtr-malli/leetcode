Given two strings s1 and s2, return the lowest ASCII sum of deleted characters to make two strings equal.

 

Example 1:

Input: s1 = "sea", s2 = "eat"
Output: 231
Explanation: Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.
Deleting "t" from "eat" adds 116 to the sum.
At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this.
Example 2:

Input: s1 = "delete", s2 = "leet"
Output: 403
Explanation: Deleting "dee" from "delete" to turn the string into "let",
adds 100[d] + 101[e] + 101[e] to the sum.
Deleting "e" from "leet" adds 101[e] to the sum.
At the end, both strings are equal to "let", and the answer is 100+101+101+101 = 403.
If instead we turned both strings into "lee" or "eet", we would get answers of 433 or 417, which are higher.
 

Constraints:

1 <= s1.length, s2.length <= 1000
s1 and s2 consist of lowercase English letters.

```python

# class Solution:
#     def minimumDeleteSum(self, s1: str, s2: str) -> int:
#         l1 = len(s1)
#         l2 = len(s2)

#         def find_remining_sum(i1, i2, total):
#             if i1 >= l1:
#                 for ele in s2[i2:]:
#                     total+= ord(ele)
#             else:
#                 for ele in s1[i1:]:
#                     total+= ord(ele)
#             return total

#         def low_sum(i1, i2, total):
#             if i1 >= l1 or i2 >= l2:
#                 return find_remining_sum(i1,i2, total)
            
#             # if the characters are equal, we can move forward.
#             if s1[i1] == s2[i2]:
#                 return low_sum(i1+1, i2+1, total)
#             # here there can be 2 things.
#             l = low_sum(i1+1, i2, total + ord(s1[i1]))
#             r = low_sum(i1, i2+1, total + ord(s2[i2]))
#             return min(l, r)
        
#         return low_sum(0,0,0)


# class Solution:
#     def minimumDeleteSum(self, s1: str, s2: str) -> int:
#         l1, l2 = len(s1), len(s2)
#         memo = [[-1] * (l2 + 1) for _ in range(l1 + 1)]

#         def remaining_sum(i1, i2):
#             total = 0
#             for c in s1[i1:]:
#                 total += ord(c)
#             for c in s2[i2:]:
#                 total += ord(c)
#             return total

#         def dp(i1, i2):
#             if memo[i1][i2] != -1:
#                 return memo[i1][i2]

#             if i1 == l1 or i2 == l2:
#                 memo[i1][i2] = remaining_sum(i1, i2)
#                 return memo[i1][i2]

#             if s1[i1] == s2[i2]:
#                 memo[i1][i2] = dp(i1 + 1, i2 + 1)
#             else:
#                 memo[i1][i2] = min(
#                     ord(s1[i1]) + dp(i1 + 1, i2),
#                     ord(s2[i2]) + dp(i1, i2 + 1)
#                 )
#             return memo[i1][i2]

#         return dp(0, 0)


class Solution:
    def minimumDeleteSum(self, s1: str, s2: str) -> int:
        l1, l2 = len(s1), len(s2)
        dp = [[-1] * (l2 + 1) for _ in range(l1 + 1)]  

        def remaining_sum(i1, i2):
            total = 0
            for c in s1[i1:]:
                total += ord(c)
            for c in s2[i2:]:
                total += ord(c)
            return total

        # base case:
        for i in range(l1):
            dp[i][l2] = remaining_sum(i, l2)
        for i in range(l2):
            dp[l1][i] = remaining_sum(l1, i)

        dp[l1][l2] = 0
        
        for i1 in range(l1-1, -1, -1):
            for i2 in range(l2-1, -1, -1):
                if s1[i1] == s2[i2]:
                    dp[i1][i2] = dp[i1 + 1][i2 + 1]
                else:
                    dp[i1][i2] = min(
                        ord(s1[i1]) + dp[i1 + 1][i2],
                        ord(s2[i2]) + dp[i1][i2 + 1]
                    )
        return dp[0][0]
```

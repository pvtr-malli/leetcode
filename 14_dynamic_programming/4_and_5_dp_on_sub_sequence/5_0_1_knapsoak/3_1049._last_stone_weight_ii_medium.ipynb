You are given an array of integers stones where stones[i] is the weight of the ith stone.

We are playing a game with the stones. On each turn, we choose any two stones and smash them together. Suppose the stones have weights x and y with x <= y. The result of this smash is:

If x == y, both stones are destroyed, and
If x != y, the stone of weight x is destroyed, and the stone of weight y has new weight y - x.
At the end of the game, there is at most one stone left.

Return the smallest possible weight of the left stone. If there are no stones left, return 0.

 

Example 1:

Input: stones = [2,7,4,1,8,1]
Output: 1
Explanation:
We can combine 2 and 4 to get 2, so the array converts to [2,7,1,8,1] then,
we can combine 7 and 8 to get 1, so the array converts to [2,1,1,1] then,
we can combine 2 and 1 to get 1, so the array converts to [1,1,1] then,
we can combine 1 and 1 to get 0, so the array converts to [1], then that's the optimal value.
Example 2:

Input: stones = [31,26,33,21,40]
Output: 5
 

Constraints:

1 <= stones.length <= 30
1 <= stones[i] <= 100


### Intuition and recursion idea:

* The problem is equivalent to partitioning stones into two subsets with minimum difference in their sums.
* Recursion tries all ways to either:

  * Include the current stone in one subset (e.g., subset1)
  * Or in the other subset (subset2)

---

### Recursion outline:

* At each stone index, you decide **which subset to put the stone into**.
* Keep track of the current sum difference (or sum of one subset).
* Recurse for next stone.
* Base case: when all stones are assigned, return the absolute difference of sums.

---

```python
from typing import List

class Solution:
    def lastStoneWeightII(self, stones: List[int]) -> int:
        def dfs(index: int, current_sum: int) -> int:
            if index == len(stones):
                return abs(current_sum)

            # Put stone in subset 1 (add weight)
            option1 = dfs(index + 1, current_sum + stones[index])
            # Put stone in subset 2 (subtract weight)
            option2 = dfs(index + 1, current_sum - stones[index])

            return min(option1, option2)

        return dfs(0, 0)

# tc - (2^n) for all subsets
# sc - O(n) for recursion stack space
```

```python
# memorization:

from typing import List

class Solution:
    def lastStoneWeightII(self, stones: List[int]) -> int:
        memo = {}

        def dfs(index: int, current_sum: int) -> int:
            if index == len(stones):
                return abs(current_sum)
            
            if (index, current_sum) in memo:
                return memo[(index, current_sum)]

            option1 = dfs(index + 1, current_sum + stones[index])
            option2 = dfs(index + 1, current_sum - stones[index])

            memo[(index, current_sum)] = min(option1, option2)
            return memo[(index, current_sum)]

        return dfs(0, 0)

# tc - (n * sum(stones)) for memoization
# sc - O(n * sum(stones)) for memoization storage
```

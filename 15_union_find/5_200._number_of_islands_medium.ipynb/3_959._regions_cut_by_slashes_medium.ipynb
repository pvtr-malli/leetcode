An n x n grid is composed of 1 x 1 squares where each 1 x 1 square consists of a '/', '\', or blank space ' '. These characters divide the square into contiguous regions.

Given the grid grid represented as a string array, return the number of regions.

Note that backslash characters are escaped, so a '\' is represented as '\\'.

 

Example 1:


Input: grid = [" /","/ "]
Output: 2
Example 2:


Input: grid = [" /","  "]
Output: 1
Example 3:


Input: grid = ["/\\","\\/"]
Output: 5
Explanation: Recall that because \ characters are escaped, "\\/" refers to \/, and "/\\" refers to /\.
 

Constraints:

n == grid.length == grid[i].length
1 <= n <= 30
grid[i][j] is either '/', '\', or ' '.

```python
from collections import deque

class Solution:
    def regionsBySlashes(self, grid: list[str]) -> int:
        n = len(grid)
        size = n * 3   # expand each cell into 3x3 grid
        expanded = [[0] * size for _ in range(size)]

        # Fill slashes in expanded grid
        for r in range(n):
            for c in range(n):
                if grid[r][c] == "/":
                    expanded[r*3][c*3+2] = 1
                    expanded[r*3+1][c*3+1] = 1
                    expanded[r*3+2][c*3] = 1
                elif grid[r][c] == "\\":
                    expanded[r*3][c*3] = 1
                    expanded[r*3+1][c*3+1] = 1
                    expanded[r*3+2][c*3+2] = 1

        def bfs(x, y):
            q = deque([(x, y)])
            expanded[x][y] = 2
            while q:
                i, j = q.popleft()
                for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:
                    ni, nj = i+dx, j+dy
                    if 0 <= ni < size and 0 <= nj < size and expanded[ni][nj] == 0:
                        expanded[ni][nj] = 2
                        q.append((ni, nj))

        # Count regions
        regions = 0
        for i in range(size):
            for j in range(size):
                if expanded[i][j] == 0:  # unvisited space
                    bfs(i, j)
                    regions += 1
        return regions
    
# Time Complexity: O((3n)^2) = O(n^2)
# Time: O(n^2) (since expanded grid is 3n Ã— 3n)
# Space: O(n^2) for queue + grid



# DUS does nto improve the time complexity,
```
